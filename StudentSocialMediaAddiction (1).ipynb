# Exploratory Analysis and Visualization

Exploratory Data Analysis(EDA) is a process of describing the data by means of statistical and visualization techniques in order to bring important aspects of that data into focus for further analysis.This involves inspecting the dataset from many angles, describing & summarizing it without making any assumptions.

Let's begin by importing matplotlib.pyplot and seaborn

```python
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
df.describe()
```

### 1. what is the Relationship between each variable in the dataset?

Let's find the correlation between each column in the dataset and we will round it to 2 decimals

```python
corr = df.corr(numeric_only = True).round(2)

corr
```

```python
#mask = np.triu(np.ones_like(corr,dtype = bool))

sns.heatmap(corr, annot = True, vmin = -1, vmax = 1, cmap = 'Reds')

plt.show()
```

### 2. what is Age of the studens who has the more Addiction score on average?

```python
plt.figure(figsize=(15,6))

highest_addiction_age = df.groupby(['Age','Academic_Level'],as_index = False)['Addicted_Score'].mean().round(2)

plt.ylabel('Addicted_Score')
plt.title('Addiction Score by Age')
sns.barplot(x = 'Age', y = 'Addicted_Score', data = highest_addiction_age,hue = 'Academic_Level',palette = 'bright')

```

### 3. what is the most used social media platform?

```python
Most_Used_Platform = df.groupby(['Age','Most_Used_Platform'], as_index = False)['Avg_Daily_Usage_Hours'].mean()
```

```python
plt.figure(figsize= (15,8))
sns.lineplot(x = 'Age', y = 'Avg_Daily_Usage_Hours', hue = 'Most_Used_Platform', data = Most_Used_Platform, style = "Most_Used_Platform")
plt.title('Most Used Platform')
```

### 4. Which gender has the more Addiction score


```python
Gender_Addiction = df.groupby('Gender',as_index =  False)['Addicted_Score'].mean(numeric_only = True)

Gender_Addiction
```

```python
sns.boxplot(x = 'Gender', y = 'Addicted_Score', data = df, palette = 'bright', legend = False)
```

### 5.Highest Recorded Avg daily Usage hours by country

```python
highest_recorded = df.groupby(['Academic_Level','Country'])['Avg_Daily_Usage_Hours'].max(numeric_only = True).sort_values(ascending = False)

highest_recorded.head()
```

```python
#Top 5 most used social media platforms
```

```python
top_platforms = df.groupby(['Most_Used_Platform'])['Avg_Daily_Usage_Hours'].mean().sort_values(ascending = False)

top_platforms.head()
```
