# Exploratory Analysis and Visualization

Exploratory Data Analysis(EDA) is a process of describing the data by means of statistical and visualization techniques in order to bring important aspects of that data into focus for further analysis.This involves inspecting the dataset from many angles, describing & summarizing it without making any assumptions.

Let's begin by importing matplotlib.pyplot and seaborn

```python
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['figure.figsize'] = (7, 3)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
df.describe()
```

# Distribution of Social Media Addiction Scores

```python
plt.figure(figsize=(10, 5))

sns.histplot(df['Addicted_Score'], bins=10, kde=True, color="steelblue")

plt.title("Distribution of Social Media Addiction Scores")

plt.show()
```

This histogram shows most students report moderate to high addiction scores, especially between 5 and 9. This supports the hypothesis that social media plays a significant role in students daily routines, potentially affecting their habits and well-being.

### 1. what is the Relationship between each variable in the dataset?

Let's find the correlation between each column in the dataset and we will round it to 2 decimals

```python
corr = df.corr(numeric_only = True).round(2)

corr
```

```python
#mask = np.triu(np.ones_like(corr,dtype = bool))

sns.heatmap(corr, annot = True, vmin = -1, vmax = 1, cmap = 'Reds')

plt.title('Correlation Heatmap')

plt.show()
```

### Addiction Score vs Mental health

```python
plt.figure(figsize=(10,6))

sns.scatterplot(data=df, x='Addicted_Score', y='Mental_Health_Score', hue='Gender')

plt.title("Addiction Score vs Mental Health Score")

plt.show()
```

The scatterplot shows a negative relationship: as addiction scores increase, reported mental health tends to decline. This supports research linking excessive screen time to poorer psychological well-being.

### 2. what is the most used social media platform?

```python
Most_Used_Platform = df.groupby(['Age','Most_Used_Platform'], as_index = False)['Avg_Daily_Usage_Hours'].count()
```

```python
plt.figure(figsize= (10,6))

sns.lineplot(x = 'Age', y = 'Avg_Daily_Usage_Hours', hue = 'Most_Used_Platform', data = Most_Used_Platform, style = "Most_Used_Platform")

plt.title('Most Used Soical Media Platform')
```

### 3. Which gender has the more Addiction score?


```python
Gender_Addiction = df.groupby('Gender',as_index =  False)['Addicted_Score'].mean(numeric_only = True)

Gender_Addiction
```

```python
plt.figure(figsize= (12,6))

sns.boxplot(x = 'Gender', y = 'Addicted_Score', data = df, palette = 'bright', legend = False)

plt.title('Gender By Addiction Score')
```

### 4. Which Academic Level has highest usage and addiction score?

```python
highest_recorded = df.groupby(['Academic_Level'])['Avg_Daily_Usage_Hours'].max(numeric_only = True).sort_values(ascending = False)

highest_recorded.head()
```

```python
plt.figure(figsize = (12,6))

sns.boxplot(x = 'Academic_Level', y = 'Addicted_Score', data = df, palette = 'bright', legend = False)

plt.title('Academic Level Impact')
```

### Relationship vs Social Media Usage

```python
plt.figure(figsize = (12,6))

sns.boxplot(x='Relationship_Status', y='Avg_Daily_Usage_Hours', data=df, palette = 'bright')

plt.title('Social Media Usage by Relationship Status', fontsize=14)

plt.show()
```

### 5.Highest Recorded Avg daily Usage hours by country ?


```python
highest_recorded = df.groupby(['Academic_Level','Country'])['Avg_Daily_Usage_Hours'].max(numeric_only = True).sort_values(ascending = False)

highest_recorded.head()
```
